<H2>Introduction</H2>

The analysis below is done on the Newman's scientific collaboration network.  The source data can be be found here: https://toreopsahl.com/datasets/#newman2001

The dataset consists of a 2 mode network consisting of one node being Authors and the other node being Papers that the Authors are associated with (i.e., were one of the writers of).  Note, the dataset merely has nodes without attributes of the actual authors and papers.) The work below projects this two mode network into two unipartite networks.  The first, with author's nodes, with edges to other author's showing a relationship where the author's have at one point worked together.  The second network would be the paper's network, showing papers as nodes, with edges to other paper's that have an author in common.

Once that projection is done, the work is then to do analysis on those unipartite networks.  That work will be described below.

Anyhow, first step below is to import the file and read into a dataframe with author and paper nodes as columns:

```python
import csv
import pandas as pd
import networkx as nx

path = "C:/school/cuny/620/proj2/"

with open(path  + "scientificCollabNetwork.txt", "r")as f:                   
    dfCollabs = pd.read_csv(f, delimiter=" ", names=("Author", "Paper")) 
```

<H3>Projection Function</H3>

Define a function that will convert (flatten) a bipartite graph into a unipartite graph.  This function is a variation of the function from HW6's sample code.

```python
def project(B,pv): 

    G = nx.Graph();
    for v in pv:
        G.add_node(v)
        for cv in B.neighbors(v):
            G.add_edges_from([(v,u) for u in B.neighbors(cv)])
    return G
```

<H3>Project the Bipartite Graph</H3>

Flatten the bipartite graphs into two unipartite graphs: authors and papers.

```python
G= nx.Graph()

for row in dfCollabs.itertuples():
    G.add_edge(row.Author, row.Paper)
    
authors =  project(G,dfCollabs.Author)    
papers = project(G, dfCollabs.Paper)
```

As we can see, when we flatten the graphs, the number of Author connections to each other, and the number of connections between papers increase dramatically.  We also see the degree of connectiveness varies widely, from a minimum of one for an author (2 for paper) up to 435.  So in a coming section we will reduce the size by using an "Islands in the Net" method.  In attempts to make smaller graphs that will be easier to analyze.

<H3>Histogram to find Potential Pruning Point</H3>

Let's run a histogram so we can perhaps see at what degree to best prune the graphs.  From the histogram, it looks like the value of 200 might be a good choice, as we can see few nodes with degree greater than that.

<H3>Pruning Method</H3>

This is variation from the Social Analytics Textbook.

```python
def prune(g, pruneLevel):
    smallG = nx.Graph()
    degree = nx.degree(g)
    for f, to in g.edges():
        if degree[f] > pruneLevel:
            smallG.add_edge(f,to)
        
    return smallG
    
```

<H3>Prune at level 200</H3>

Try pruning at level 200 we examined in the histograms.  When we do that we see the graph is still extremely connected, as the degrees are down to 432 for both, from 435, and there are no separate connected_compent graphs.  In the next section we will attempt Islands in the net to prune further.

```python
authorSubset = prune(authors, 200)
paperSubset = prune(papers, 200)

for c in nx.connected_components(authorSubset):
    print ("Author Number of Nodes", str(len(c)))
       

for c in nx.connected_components(paperSubset):
    print ("Paper Number of Nodes", str(len(c)))       
```

<H3>Islands in the Net Method</H3>
A method to implement islands in the net.  Again a variation from the Social Analytics Text.  This method though uses degrees of nodes to make islands, rather than weights, as this is an weighted graph.  It calls the prune method from  above.

```python
def getIslandGraphs(g, iterations):
    degrees = dict(g.degree()).values()
    mn = min(degrees)
    mx = max(degrees)
    step = int((mx-mn)/iterations)
    return [[threshold, prune(g, threshold)] for threshold in range(mn, mx, step)]
```

<H3>Execute Islands in the Net</H3>

Call the Islands in the Net method to give us an array of graphs, and their prune point based on degree.

```python
authorIslands = []
paperIslands = []

authorIslands = getIslandGraphs(authors, 10)
paperIslands = getIslandGraphs(papers, 10)
```

<H3>Examine Islands in the Nets Results </H3>

Printing our results, shows that running the island method, indeed prunes our graphs, but does not create a series of Islands as we might have hoped.  What is shows us, is that everything is connected and we don't have a nodes that can be seperated from the main graph.  At most with the collection of subsetted or pruned Authors' graphs, there is only one separated (and lonely) Author who is not connected, but everywhere else, all is connected in both sets of graphs.

```python
for g in authorIslands:
    for c in nx.connected_components(g[1]):
        print ("Author Degree Prune Point: ", str(g[0]), "Number of Nodes", str(len(c)))
        
for g in paperIslands:
    for c in nx.connected_components(g[1]):
        print ("Paper Degree Prune Point: ", str(g[0]), "Number of Nodes", str(len(c)))        
```
